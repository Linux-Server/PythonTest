```python
import pandas as pd

data = pd.read_csv("../csv/melb_data.csv")
data.head()
```

```python
from sklearn.model_selection import train_test_split
y = data.Price
X = data.drop(["Price"], axis=1)
X = X.select_dtypes(exclude=["object"])
X_train, X_valid, y_train, y_valid = train_test_split(X,y,train_size=0.8, test_size=0.2, random_state=0)

X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def score_datset(X_train,X_valid,y_train,y_valid):
    model = RandomForestRegressor(n_estimators=10, random_state=0)
    model.fit(X_train,y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid,preds)
    
```

```python
result = score_datset(X_train,X_valid,y_train,y_valid)
print(result)

X_train.columns
```

```python
missing_train_data = [col for col in X_train.columns
                       if X_train[col].isnull().any()]
```

```python
missing_train_data
```

```python
reduced_X_train = X_train.drop(missing_train_data, axis=1)
reduced_X_valid = X_valid.drop(missing_train_data, axis=1)
```

```python
reduced_X_train.columns
```

```python
res= score_datset(reduced_X_train,reduced_X_valid,y_train,y_valid)
```

```python
res
```

```python
from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()

my_imputer

X_valid.iloc[0]
```

```python
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))
imputed_X_valid.columns = X_valid.columns
imputed_X_train.columns = X_train.columns
imputed_X_valid
```

```python
score_datset(imputed_X_train,imputed_X_valid,y_train,y_valid)
```

```python
import pandas as pd
data = pd.read_csv("../csv/melb_data.csv")
data.head()
```

```python
y= data.Price
y
```

```python
melb_preditions = data.drop(["Price"], axis=1)
X = melb_preditions.select_dtypes(exclude=["object"])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_valid,y_train,y_valid = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def score_dataset(X_train,X_valid,y_train,y_valid):
    model = RandomForestRegressor(n_estimators=10, random_state=0)
    model.fit(X_train,y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
    
```

```python
col_with_missing = [col for col in X_train.columns
                    if X_train[col].isnull().any()]

col_with_missing
```

```python
reduced_X_train = X_train.drop(col_with_missing, axis=1)
reduced_X_valid = X_valid.drop(col_with_missing, axis=1)
```

```python
print(score_dataset(reduced_X_train, reduced_X_valid, y_train, y_valid))
```

```python
X_train.columns
```

```python
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer()
X_train
```

```python
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))

# Imputation removed column names; put them back
imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns

print("MAE from Approach 2 (Imputation):")
print(score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid))
```
